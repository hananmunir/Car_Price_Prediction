```python
#import libraries
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV
import pickle
import matplotlib.pyplot as plt
%matplotlib inline

# importing date class from datetime module
from datetime import date
```

```python
#read dataset
df = pd.read_csv('car data.csv')
```

```python
#updating dataset
final_dataset = df[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven',
       'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
```

```python
#dataset head
final_dataset.head()
```

```python
# creating the date object of today's date
todays_date = date.today()
    
#adding a new attribute
final_dataset['Current_Year'] = todays_date.year
```

```python
#dataset head
final_dataset.head()
```

```python
#adding no of years attribute
final_dataset['no_year'] = final_dataset['Current_Year'] - final_dataset['Year']
```

```python
#dataset head
final_dataset.head()
```

```python
#dropping unneccessary attributes
final_dataset.drop(['Year'],axis = 1, inplace = True)
final_dataset.drop(['Current_Year'],axis = 1, inplace = True)
```

```python
#dataset head
final_dataset.head()
```

```python
#Create Dummy Data
final_dataset = pd.get_dummies(final_dataset, drop_first = True)
```

```python
#dataset head
final_dataset.head()
```

```python
#getting correlation
final_dataset.corr()
```

```python
#generating correlation diagrams
sns.pairplot(final_dataset)
```

```python
corrmat = final_dataset.corr() 
top_corr_features = corrmat.index 
plt.figure(figsize = (40,40)) 
#plot heat map 
g = sns.heatmap(final_dataset[top_corr_features].corr(), annot = True, cmap = "RdYlGn")
```

```python
#independent features
x = final_dataset.iloc[ : , 1 : ]
#dependent feature
y = final_dataset.iloc[ : , 0]
```

```python
x.head()
```

```python
y.head()
```

```python
# prioritizing features
model = ExtraTreesRegressor()
model.fit(x,y)
```

```python
print(model.feature_importances_)
```

```python
# getting top five most important features
feat_importance = pd.Series(model.feature_importances_, index = x.columns)
feat_importance.nlargest(5).plot( kind = 'barh')
plt.show()
```

```python
#train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .2) 
```

```python
#printing shapes
print("Shape of Training Data of X", x_train.shape)
print("Shape of Testing Data of X", x_test.shape)
print("Shape of Training Data of Y", y_train.shape)
print("Shape of Testing Data of Y", y_test.shape)
```

```python
# Randomized Search CV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]

# Number of features to consider at every split
max_features = ['auto', 'sqrt']

# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]

# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]

# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
# create random grid
random_grid = {
    'n_estimators' : n_estimators,
    'max_features' : max_features,
    'max_depth' : max_depth,
    'min_samples_split' : min_samples_split,
    'min_samples_leaf' : min_samples_leaf
}
```

```python
# use random grid to search for best hyperparameters 
rf = RandomForestRegressor()
```

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, scoring = 'neg_mean_squared_error', n_iter = 10, cv = 5, verbose = 2, random_state = 42, n_jobs = 1 )
```

```python
#train model
rf_random.fit(x_train,y_train )
```

```python
#prediction
predictions = rf_random.predict(x_test)
predictions
```

```python
sns.distplot(y_test-predictions)
```

```python
plt.scatter(y_test, predictions)
```

```python
file = open('random_forest_regression_model.pkl', 'wb')

pickle.dump(rf_random, file)
```
